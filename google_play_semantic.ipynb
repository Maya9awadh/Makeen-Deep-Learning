```python
import pandas as pd
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
import tensorflow as tf
```

```python
df = pd.read_csv("googleplaystore_user_reviews.csv")
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
df.columns
```

```python
df1=df[['Translated_Review','Sentiment']].copy()
```

```python
df1['Sentiment'].unique()
```

```python
df1.head()
```

```python
max_features=1000
```

```python
#tokenizer
tokenizer= Tokenizer(num_words=max_features,split=' ')
```

```python
tokenizer.fit_on_texts(df1['Translated_Review'].values)
```

```python
x=tokenizer.texts_to_sequences(df1['Translated_Review'].values)
```

```python
tokenizer.index_word
```

```python
[i.__len__() for i in x[:10]]
```

```python
x[0].__len__(), df1.iloc[0]["Translated_Review"].split(" ").__len__()
```

```python
x=pad_sequences(x)
```

```python
[i.__len__() for i in x[:10]]
```

```python
y=pd.get_dummies(df1["Sentiment"]).values
```

```python
dataset = tf.data.Dataset.from_tensor_slices((x, y))
```

```python
list(dataset.take(1).as_numpy_iterator())
```

```python
for i in dataset.take(1):
    break
```

```python
i[0].shape
```

```python
dataset = dataset.batch(32)
len(list(dataset))
```

```python
train_dataset = dataset.take(300)
test_dataset = dataset.skip(300)
```

```python
len(list(train_dataset)), len(list(test_dataset))
```

```python
# make model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=[194]))
model.add(tf.keras.layers.Embedding(
    input_dim=tokenizer.word_counts.__len__(),
    output_dim=64,
    input_length=29
))
model.add(tf.keras.layers.LSTM(200, return_sequences=True))
model.add(tf.keras.layers.LSTM(100))
model.add(tf.keras.layers.Dense(3, activation="softmax"))
```

```python
model.summary()
```

```python
cb = [
    tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=5),
    tf.keras.callbacks.TensorBoard(log_dir="logs", histogram_freq=1, write_graph=True, embeddings_freq=1, embeddings_metadata="./logs")
]
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss=tf.keras.losses.categorical_crossentropy)
```

```python
model.fit(dataset, 
          validation_data=test_dataset, 
          epochs=1000, 
          callbacks=cb)
```
